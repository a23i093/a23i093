(2)2つのリストの要素をたす。<br>
numpyではない

```python
L5=[2, 3, 4]
L6=[1, 7, 9]

#後ろに追加されるだけ。
L7 = L5+L6
print("L7:", L7)

#forでつくる
L8=[]
for (i,j) in zip(L5,L6):
  L8.append(i+j)
print("L8:", L8)

L9=[i+j for (i,j) in zip(L5,L6)]
print("L9:", L9)
```

#3.numpyで行列の計算(AIは行列計算の塊）

(1)行列の演算(+,べき乗)

```python
#2次元リスト　3x3の行列
L1=[[1,3,6],[2,6,12],[3,9,18]]

#numpyで行列に変換
M1=np.array(L1)

#行列の足し算
M2 = M1+M1

print("M2:\n", M2)

#行列のべき乗
M3 = M2**3

print("M3:\n", M3)
```
