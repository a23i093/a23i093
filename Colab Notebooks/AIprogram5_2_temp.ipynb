(3)関数の引数が複数(ここでは２つ)ある場合

```python
def p2(n, msg):
  print(msg)
  for _ in range(n):
    print("リンゴも食べたい")

p2(5, "リンゴの出力の関数5回")
p2(2, "リンゴの出力の関数2回")
```

(4)関数の引数にリストを使う

```python
def p3(msg, f_l):
  print(msg,"の表示")
  for f in f_l:
    print("{}も食べたい".format(f))

fruits_list=["リンゴ", "メロン", "ミカン"]

p3("果物", fruits_list)
```

(5)関数の引数に複数のリストを使う

```python
def p4(msg, f_l, f_p):
  print(msg, "の表示")
  for n,v in zip(f_l, f_p):
    print("{}円の{}が食べたい".format(v, n))

fruits_list=["リンゴ", "メロン", "ミカン"]
fruits_price=[150, 780, 100]

p4("果物の価格", fruits_list, fruits_price)
```

# 課題4

xとyの値は共に0以上である場合は、正常入力ですを出力し、<br>
そうでない場合は、何も出力しないように、<br>
関数ex_funcを作りなさい。

```python
def ex_func1(a,b):
  if a>=0 and b>=0:
    return(1)
  else:
    return(0)

x = int(input("変数xに入れる整数を入力してください"))
y = int(input("変数yに入れる整数を入力してください"))

if ex_func1(x,y) == 1:
  print("正常入力です")
else:
  print("異常入力です")
```

(8)関数の返却にリストを使う<br>
グラフ表示をしてみる。matplotlibを使って。

```python
#matplotlibを使うための準備
import matplotlib.pyplot as plt

def func(x):
  return 2*x**2-3

def func1(x):
  return [func(x1) for x1 in x]

x1=[x for x in range(-10,10,1)]
y1=[func(x) for x in x1]


print("x1=", type(x1),x1)
print("y1=",type(y1),y1)

#グラフで表示
#この場合のplot関数の第一引数x1はx軸に対応し、
#第二引数のyがy軸にあたります。
plt.plot(x1,y1)
plt.show()
```

#課題7

a=?,b=?,c=?,d=?を決め、ax^3+bx^-2+cx+dのグラフを描きなさい。

```python
import matplotlib.pyplot as plt

def func(x, a, b, c, d):
  return a*x**3+b*x**2+c*x+d

def func1(x, a, b, c, d):
  return [func(x1, a, b, c, d) for x1 in x]

#a,b,c,dの設定
a=3
b=-6
c=5
d=-3

x1=[x for x in range(-10,10,1)]
y1 = func1(x1, a, b, c, d)

print("x1=", type(x1),x1)
print("y1=",type(y1),y1)

# この場合のplot関数の第一引数x1は、x軸に対応し、
#第二引数のy1がy軸にあたります。
plt.plot(x1,y1)
plt.show()
```
