課題8<br>
3のデータa,b,cを入力したときに、最小値を出力する<br>
プログラムを作成せよ。<br>
論理演算andを使い、ifの中にifは使わないで書く。<br>


````python
#課題8のプログラムをここに書く
a=int(input())
b=int(input())
c=int(input())
if (a<b)and(a<c):
  min=a
elif (b<a)and(b<c):
  min=b
else:
  min=c
print("最小値は、{}です。".format(min))
````

課題9 入力をscoreとし、scoreの範囲が、<br>
0より小さい、または100より大きなら"error"<br>
0以上60未満なら、"不合格"<br>
60以上100なら、"合格"<br>
を出力するプログラムを作成しなさい。


````python
#課題9のプログラムをここに書く
score = int(input())
if(score<0)or(score>100):
  print("error")
elif(score>=0)and(score<60):
  print("不合格")
elif(score>=60)and(score<=100):
  print("合格")

````
