2.3 クラスの継承　メソッドのオーバーライトと親のメソッドを呼び出し

```
class Nigiri:
  category ="にぎり"
  top = "ねた"
  base = "しゃり"

  def show_attribute(self):
    print("top:{}, base:{}, category:{}".format(self.top,self.base,self.category))

class Maguro(Nigiri):
  top = "まぐろ"
  price = 200
  parts = "とろ"

  def process(self):
    self.top = self.top + "("+self.top+")"

  def show_attribute(self):
    return super().show_attribute()

  def show_price(self, num = 2):
    result = self.price * num
    print("個数{}, お値段{}".format(num, result))

n1 = Nigiri()
n1.show_attribute()

m1 = Maguro()
m1.process()
m1.show_attribute()
m1.show_price()
```

課題9　問題のプログラムを入力して、結果を確かめなさい。

```
class Car:
  def __init__(self, model, color):
    self.model = model
    self.color = color

  def display(self):
    print('model:', self.model)
    print('color:', self.color)


class Truck(Car):
  def __init__(self,model,color,cargo):
    Car.__init__(self,model,color)
    self.cargo = cargo

  def display(self):
    Car.display(self)
    print('cargo :', self.cargo)

  def load(self, cargo):
    self.cargo = cargo

truck1 = Truck('trailer', 'yellow', 'potato')
truck1.load('lettuce')
truck1.display()
```
