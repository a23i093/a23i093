#課題2<br>


```
my_str = 'abcあいうえ'
```
としたときに、my_strを分割して
変数my_str3に'abc'の部分、my_str4に'あいうえ'の部分に分割してそれぞれを出力するプログラムを作成せよ。


````python
#課題2のプログラムをここに書く
my_str = 'abcあいうえ'
my_str3 = my_str[0:3]
my_str4 = my_str[3:7]
print("my_str3 = ",my_str3)
print("my_str4 = ",my_str4)
````
