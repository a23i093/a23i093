```python
D1={"a":3, "b":-9, "c":1, "d":-6}

def abs_d(dictionary):
    for key,value in dictionary.items():
        if value < 0:
            dictionary[key] *= -1
    return(dictionary)

D2 = abs_d(D1)

print("D2=", D2)
```
