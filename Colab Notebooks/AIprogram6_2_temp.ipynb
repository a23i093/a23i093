1.5 初期化メソッド

```
class Car: #Carクラスを作成
  num_wheels = 4
  weight = 3000
# name = "ある"

  def __init__(self,car_name = "なんだろう"): #インスタンス生成時の実行
    self.name = car_name

  def cal_weight_per_wheel(self):  #メソッドの作成 selfでクラス変数にアクセス
    self.name = "あれ？"
    return("車輪後ろの重さ", self.weight/self.num_wheels)

default_car = Car() #インスタンスをつくる。引数無し
print ("default_carの名前:", default_car.name)

papa_car=Car("ファミリーカー")#インスタンスをつくる。引数あり
print ("papa_carの名前:", papa_car.name)

mama_car=Car("軽")#インスタンスをつくる。引数あり
res=mama_car.cal_weight_per_wheel()#メソッドを実行
print ("mama_carの名前:", mama_car.name)
print ("mama_carでのresの内容:", res)
```

問題2<br>
インスタンスの生成時に、引数を指定し、<br>
そのインスタンスに対するcal_plusとcal_minusで<br>
足し算と引き算を計算するプログラムを作成せよ。

```
#問題2
class Calc:
  def __init__(self, x=0 , y=0):
    self.x = x
    self.y = y

  def cal_plus(self):
    return self.x + self.y

  def cal_minus(self):
    return self.x - self.y

data3 = Calc()
print("計算1",data3.cal_plus())
print("計算1",data3.cal_minus())

data4=Calc(2,5)
print("計算2",data4.cal_plus())
print("計算2",data4.cal_minus())
```

課題7　問題のプログラムを入力して、結果を確かめなさい。

```
class Car:

  def __init__(self, model, color):
    self.model = model
    self.color = color
    self.speed = 0

  def display(self):
    print("model:", self.model)
    print("color:", self.color)
    print("speed:", self.speed)

  def accelerate(self, pressure):
    self.speed += pressure
    self.__check_speed()

  def __check_speed(self):
    if self.speed > 100:
      self.speed = 100

car1 = Car("roadster", "red")
car1.display()
car1.accelerate(30)
car1.accelerate(50)
car1.display()
```
