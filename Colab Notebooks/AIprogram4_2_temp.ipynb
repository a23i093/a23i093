課題7<br>
名前と得点の組からなる辞書データのうち、最高得点の名前とそのスコアを出力する<br>
プログラムを作成せよ。

```python
dict2 = {"asa":80, "yan":90, "shin":50}

d_max=-float('inf')

#forとifで、最大となる値をd_max・最大となる値のkeyをk_maxに入れる。
for key,value in dict2.items():
  if d_max < value:
    d_max = value
    k_max=key
print("最大値のkeyは{}、最大値は{}です。".format(k_max, d_max))
```

課題8<br>
名前と得点の組からなる辞書データのうち、得点について分散を<br>
出力するプログラムを作成せよ。<br>
 image.png

```python
dict2 = {"asa":80, "yan":90, "shin":50}

#平均の計算のための合計
d_sum = 0
for key,value in dict2.items():
  d_sum += value
d_ave=d_sum/len(dict2)

#分散の計算varianceに格納
variance=0
for k,v in dict2.items():
  variance += (v-d_ave)**2
variance /= len(dict2)
print("分散=", variance)
```
