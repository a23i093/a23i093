# リスト内包表記

(1)リストの中でプログラムを書いてリストをつくる。

```python
data = [3, 1, -4, 5, -7]
print("data=", data)

#リスト内包表記[]の中にプログラム
ndata=[2*x+3 for x in data]
print("ndata=",ndata)
```

(2)リスト内包表記条件付き

```python
data = [3, 1, -4, 5, -7]
print("data=", data)

#リスト内包表記[]の中にプログラム
ndata=[2*x+3 for x in data if x > 0]
print("ndata=",ndata)
```

(3)リスト内表表記で指定個数のリストを作る

```python
n=10

pdata=[i for i in range(n)]

print("pdata=",pdata)
```

(4)リスト内包表記で奇数のデータをn個作る

```python
import random
n = 10

pdata=[random.randrange(-9, +9, 2) for i in range(n)]
print("pdata=",pdata)
```

課題1<br>

リスト内包表記を使ってリストを作り、求めなさい。

```python
data = [3, 1, -4, 5, -7, 3, -9]
print("data=", data)

#リスト内包表記[]の中にプログラム
ndata=[2*x+3 for x in data if x < 0]
print("負のデータの合計:",sum(ndata))
```

課題4<br>
リスト内包表記を使って、-100～100の間の乱数をn個作り、<br>
リスト内の正の総和と負の総和を出力するプログラムを作成せよ。<br>
なお、作成した乱数のリストをforで処理して和を求めること。

```python
n=10
pdata=[random.randint(-100,100) for i in range(n)]
print("sum pdata=",pdata)

p_sum=0
n_sum=0

for i in pdata:
  if i > 0:
    p_sum += i
  elif i < 0:
    n_sum += i

print("p_sum=", p_sum)
print("n_sum=", n_sum)
```

課題5<br>
リスト内包表記を使って、-100～100の間の乱数をn個作り、<br>
リスト内の正の総和と負の総和を出力するプログラムを作成せよ。<br>
なお、正のリストと、0以下のリストを内包表記で作成し、<br>
リストの合計には関数sumを用いること。

```python
n=10

pdata=[random.randint(-10,10) for i in range(n)]
print("pdata=",pdata)

p_list=[x for x in pdata if x > 0]
print("p_list=", p_list)

n_list=[x for x in pdata if x < 0]
print("n_list=", n_list)

print("p_sum=", sum(p_list))
print("n_sum=", sum(n_list))
```

課題6(参考用)<br>
-10から10までの整数の乱数をn個作りリストとする。<br>
リストの0以上の個数と0未満の個数を出力するプログラムを<br>
作製せよ。<br>
下記プログラムは、正と負のリストはforを用いている。

```python
n=10

pdata=[random.randint(-10,10) for i in range(n)]
print("pdata=",pdata)

p_list=[]
n_list=[]

for i in pdata:
  if i > 0:
    p_list.append(i)
print("p list=",p_list)


for i in pdata:
  if i <= 0:
    n_list.append(i)
print("n list=",n_list)

p_num = 0
for i in p_list:
  p_num += 1

n_num = 0
for i in n_list:
  n_num += 1

print("0以上の個数=", p_num)
print("0未満の個数=", n_num)
```

課題6 作成用<br>
-10から10までの整数の乱数をn個作りリストとする。<br>
リストの0以上の個数と0未満の個数を出力するプログラムを作製せよ。<br>
なお、リスト内包表記で0以上のデータのリストと<br>
0未満のデータのリストに分け、<br>
len関数を使って、それぞれのリストの数を求める。


```python
n=10

pdata=[random.randint(-10,10) for i in range(n)]
print("pdata=",pdata)

p_list=[i for i in pdata if i >= 0]
print("p_list=", p_list)

n_list=[i for i in pdata if i < 0]
print("n_list=", n_list)

print("0以上の個数=", len(p_list))
print("0未満の個数=", len(n_list))
```
