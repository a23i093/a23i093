#複雑な関数

(9)関数の中から関数の中にある関数を呼び出す

```
import matplotlib.pyplot as plt

#関数内関数
def func1(x):
  def funcx(x):
    return 2*x**2-3

  return[funcx(x1) for x1 in x]

x1=[x for x in range(-10,10,1)]
y1=func1(x1)

print("x1=", type(x1),x1)
print("y1=",type(y1),y1)

# この場合のplot関数の第一引数x1は、x軸に対応し、
#第二引数のy1がy軸にあたります。
plt.plot(x1,y1)
plt.show()
```

#課題8

y=ax^2+bx+cのa,b,cを入力し、<br>
グラフを描くプログラムを作成せよ。<br>
プログラムは、関数（関数内関数)を使ってy1のリストを作り<br>
最後にpyplotでグラフを描く

```
import matplotlib.pyplot as plt
#関数内関数
def func2(x, a, b, c):
  def funcx(x, a, b, c):
    return a*x**2+b*x+c

  return [funcx(x1, a, b, c) for x1 in x]

#ax^2+bx+cの係数の入力
a=float(input("係数aの入力"))
b=float(input("係数bの入力"))
c=float(input("係数cの入力"))

#x1とそれに対応するy1のリストの計算
x1=[x for x in range(-10,10,1)]
y1=func2(x1,a,b,c)


print("x1=",x1)
print("y1=",y1)

plt.plot(x1,y1)
plt.show()
```

(10)可変引数

```
def fruits_name2(*args):
  print("args:", args)
  for arg in args:
    print(arg)

# 引数の違いによる関数の違い
print("関数fruite_name2の呼び出し1 引数が2")
fruits_name2("apple", "リンゴ")#結果の違いを確認
print("関数fruite_name2の呼び出し2 引数が3")
fruits_name2("apple", "リンゴ", "orwnge")
```

(11)可変引数の呼び出し側に変数を使う

```
def fruits_name2(*args):
  print("args:", args)
  for arg in args:
    print(arg)

n_t1 = ("apple", "orange")  # タプル型のデータの代入
n_t2 = ("apple", "orange", "abc")  #

#2つ目 変数にするとうまく取れない
print("関数fruite_name2の呼び出し3 引数がタップル")
fruits_name2(n_t1)
fruits_name2(n_t2)

print(" ")
#3つ目 変数の場合、ちょっと工夫を*をつける
print("関数fruite_name2の呼び出し4 引数がタップル")
fruits_name2(*n_t1)
fruits_name2(*n_t2)
```

(12)可変引数の呼び出し側に変数を使う　その２

```
def fruits_name2(*args):#複数の引数を扱う関数
  print("args:", args)  #引数の表示
  for arg in args:      #引数の各要素の表示
    print(arg)

n_t3=["a", "b"]
print("関数fruite_name2の呼び出し5 引数がリスト")
fruits_name2(n_t3)
fruits_name2(*n_t3)

n_t4 = {"a":123, "c":321}
print("関数fruite_name2の呼び出し6 引数が辞書")
fruits_name2(n_t4)
fruits_name2(*n_t4)
```

参考）多項式のグラフを描く

```
import matplotlib.pyplot as plt
#関数内関数
def func(x,*args):
  def funcx(x,*coef):
    f=0
    for i,a in enumerate(coef):
      if i==0:
        f=a
#        print("f={}a={},x={},i={}".format(f,a,x,i))
      else:
        f = f*x+a
#        print("f={}a={},x={},i={}".format(f,a,x,i))
    return f

  return [funcx(x1,*args) for x1 in x]

#ax^4+bx^3+cx^2+dx+eの係数の入力
coef=(1,-2,1,1,1)

x1=[x for x in range(-10,11,1)]
y1=func(x1,*coef)

plt.plot(x1,y1)
plt.show()
```
