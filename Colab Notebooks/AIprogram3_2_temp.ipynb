課題3<br>
リストの逆順を出力するプログラムを、whileを使って作りなさい。

```python
mylist_e = ["banana", "melon", "orange", "apple"]

n=len(mylist_e)
i=n-1

while i >= 0:
  print("{}番目の項目{}".format(i,mylist_e[i]))
  i -= 1
```

# 2.7リスト(3)

(1)リストの操作

```python
mylist_e = ["banana", "melon", "orange", "apple"]

fruit = input()

mylist_e.append(fruit)
print("追加されたリスト",mylist_e)
```

(2)途中に追加 insert

```python
mylist_e = ["banana", "melon", "orange", "apple"]

fruit = input("置き換える果物を入力:")

mylist_e.insert(2, fruit)
print("追加されたリスト",mylist_e)
```

(3)リストの中の先頭から一致するものを削除する

```python
mylist_e = ["banana", "melon", "orange", "apple"]

fruit = input("削除した果物")

mylist_e.remove(fruit)
print("削除後のされたリスト",mylist_e)
```

(5)リスト中の特定の場所の要素を削除する(pop)<br>
アルゴリズム,データ構造ではpopはスタックの<br>
先頭を取り出します。プログラムなの少し違う。


```python
mylist_e = ["banana", "melon", "orange", "apple"]

d_fruit = input("削除する果物を指定してください(英語)。")

print("{}番目の{}を削除します".format(mylist_e.index(d_fruit),d_fruit))

d_index = mylist_e.index(d_fruit)

mylist_e.pop(d_index)
print("mylist_e=", mylist_e)
```

課題4 <br>
リストの操作<br>
英語の果物を指定し1つ削除し、同様に日本語の果物名も削除する

```python
mylist_e = ["banana", "melon", "orange", "apple"]
mylist_j = ["バナナ", "メロン", "オレンジ", "リンゴ"]

d_fruit = input("削除する果物を指定してください(英語)。")

print("i= ",mylist_e.index(d_fruit))
d_index = mylist_e.index(d_fruit)

mylist_e.pop(d_index)
mylist_j.pop(d_index)

print("mylist_e=",mylist_e)
print("mylist_j=",mylist_j)
```

課題5<br>
英語のリストに対して、日本語のリストを入力し、最後に日本語のリストを出力する

```python
mylist = ["banana", "melon", "orange", "apple"]
mylist_j = []

for fruit in mylist:
    j_fruit = input("{}の日本語を入力してください。".format(fruit))
    mylist_j.append(j_fruit)

print("果物の日本語のリスト;",mylist_j)

```
