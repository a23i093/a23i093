# 3.3 for文

(1)単純にn回繰り返す<br>
```
for　i in range(n):
で、range(n)で0～n-1までiに1ずつ増やして
繰り返す。
range(1,n)で0～n-1までiに1ずつ増やす。
range(1,n+1)で0～nまでiに1ずつ増やす。
range(1,n,2)で0～n-1までiに2ずつ増やす。
```
下のプログラムは、nまでちゃんと足せていないので間違いです。



````python
n = int(input())

sum=0
for i in range(n):
  sum += i
print("{}の総和={}".format(n,sum))
````

課題10<br>
入力nの奇数の総和と、偶数の総和を<br>
出力するプログラムを作成せよ。

````python
#課題10のプログラムをここに書く
n = int(input())

sum_o=0
sum_e=0

for i in range(1,n+1,2):
    sum_o += i
for i in range(2,n+1,2):
  sum_e +=i

print("奇数の総和=", sum_o)
print("偶数の総和=", sum_e)
````

(2)リストから繰り返しを実行する。<br>
forの中にifを書いた場合

````python
sum_e=0
for i in [1,3,7,6,2]:
  if i % 2 == 0:
    sum_e += i
print("リストの偶数の総和=",sum_e)
````

課題12
リストの正の数の合計を出力するプログラムを作成せよ。

````python
#課題12のプログラムをここに書く
sum_p=0

for i in [1, -3, 7, 8, 0.1, 7, -20]:
  if (i>0):
    sum_p += i
print("0以上の合計=", sum_p)
````
