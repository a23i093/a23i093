(2)nまでの総和(while版)

````python
n=10
i=0
sum=0
while i<=n:
  sum += i
  i += 1
print("{}までの総和は{}です。".format(n,sum))
````

(3)特定の入力が来るまで繰り返す

````python
str = input("文字を入力(qで終了)")

while str != "q":
  print("入力文字", str)
  str = input("文字を入力(qで終了)")
print("whileが終了しました")
````

課題1<br>
正の整数を入力し、負の値をいれたら、それまでの合計を出力する<br>
プログラムを作成せよ。

````python
n = int(input("整数を入力(負で終了):"))
sum = 0

while n >= 0:
  print("入力", n)
  sum += n
  n = int(input("整数を入力(負で終了):"))
print("合計は{}です。".format(sum))
````

# 2.7 リスト

(1)リストとfor

````python
list = [1, 5, 7, 6, -3]
for i in list:
  print("i=", i)
````

(2)リストを使ったfor＋if

````python
list = [1, 5, 7, 6, -3]
sum=0
for i in list:
  if (i > 0):
    sum += i
print("sum=", sum)
````

(3)文字列のリストとfor

````python
mylist = ["banana", "melon", "orange", "apple"]
for i in mylist:
  print("i=", i)
````

(4)キーボードからの入力とリストの判定

````python
mylist_e = ["banana", "melon", "orange", "apple", "muscat", "blueberry"]
s = input()

if s in mylist_e:
  print ("yes")
else:
  print("No")
````
