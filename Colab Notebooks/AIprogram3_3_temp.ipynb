# リストとfor(組み合わせていろいろな処理に)

(1)リストを文字列に（無理にforと文字列の＋を使う）

```python
mylist = ["banana", "melon", "orange", "apple"]

mylist_string=""

for i in mylist:
  mylist_string=mylist_string + i+" "
  print("i=",i)
print("mylist_string=",mylist_string)
```

(2)リストの文字列のメソッド（countメソッド)

```python
mylist = ["banana", "melon", "orange", "apple"]
mylist_string=""
for i in mylist:
  mylist_string=mylist_string + i+" "
  print('{}の中の{}の数={}'.format(i,'a',i.count("a")))
print("myliat_string=",mylist_string)
```

(3)頭文字を大文字に変えるプログラム<br>
文字列の変数.titleで、頭文字を大文字に

```python
mylist1 = ["banana", "melon", "orange", "apple"]
mylist2 = []

for i in mylist1:
  print("L:",i.title())

  mylist2.append(i.title())
print("mylist2",mylist2)
```

課題6<br>
リストの文字列の母音の数を調べるプログラムを作成せよ。<br>
forの2重ループになる

```python
mylist = ["banana", "melon", "orange", "apple"]
vowels = ["a", "i", "u", "e", "o"]
mylist_v=[]

for fruit in mylist:
  vn = 0
  for i in vowels:
    vn += fruit.count(i)
  mylist_v.append(vn)

print("mylist_v=",mylist_v)
```

課題7<br>
新しい果物の英語と日本語を入力し、<br>
最後に出来上がっているリストを出力する<br>
プログラムを作成せよ。

```python
mylist_e = ["banana", "melon", "orange", "apple"]
mylist_j = ["バナナ", "メロン", "オレンジ", "リンゴ"]

new_fe = input("追加する果物(英語):")
new_fj = input("追加する果物(日本語):")

mylist_e.append(new_fe)
mylist_j.append(new_fj)

for e, j in zip(mylist_e, mylist_j):
  print("英語{}, 日本語{}".format(e, j))
```
