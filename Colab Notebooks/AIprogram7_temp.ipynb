```
class student:

  def set_name(self, arg1):
    if arg1 != "":
      self.name = arg1

  def set_number(self, arg1):
    self.number = arg1

  def disp(self):
    print("学生番号:{},氏名:{}".format(self.number, self.name))

#インスタンスを作る(shinをつくる)
shin = student()
#いきなり出力してみる（エラーになる)
#shin.disp()

#名前のセット
shin.set_name("しん")
#学生番号のセット
shin.set_number("0001")

#表示dispメソッドを
shin.disp()
```

```
class Student:
    def __init__(self, arg1="", arg2=0):
        self.name = arg1
        self.number = arg2
        self.total_unit = 0

    def set_name(self, arg1):
        if arg1 != "":
            self.name = arg1

    def set_unit(self, arg1):
        self.total_unit = arg1
        return(self.total_unit)

    def set_number(self, arg1):
        self.number = arg1

    def disp(self):
        print("学生番号:{}, 氏名:{}".format(self.number, self.name))
        print("総単位:{}".format(self.total_unit))

shin = Student()
shin.disp()
shin.set_name("しん")
shin.set_number("0001")
shin.set_unit(50)

shin.disp()
```

student クラスの設計4<br>
テスト評価による単位を追加とクラス変数

```
class student:
  SCHOOL_NAME = "OUS"

  def __init__(self, arg1="", arg2=0):
    self.name=arg1
    self.number=arg2
    self.total_unit=0

  def set_name(self, arg1):
    if arg1 != "":
      self.name = arg1

  def set_unit(self, arg1):
    self.total_unit = arg1
    return(self.total_unit)

  def set_number(self, arg1):
    self.number = arg1

  def evaluate_test(self, arg1):
    if arg1 >= 60:
      self.total_unit += 2
    return(self.total_unit)

  def disp(self):
    print("学生番号:{},氏名:{}".format(self.number, self.name))
    print("総単位:{}".format(self.total_unit))

shin = student()
shin.disp()
shin.set_name("しん")
shin.set_number("0001")
shin.set_unit(50)
shin.evaluate_test(90)

shin.disp()
print("shinさんの学校名",shin.SCHOOL_NAME)

yan = student("やん")
yan.set_unit(90)
yan.disp()

print("yanさんの学校名", yan.SCHOOL_NAME)

#クラス変数の変更
student.SCHOOL_NAME = "岡山理科大学"
print("yanさんの学校名", yan.SCHOOL_NAME)
print("shinさんの学校名", shin.SCHOOL_NAME)
```

# 課題1

```
class student:
  SCHOOL_NAME = "OUS"

  def __init__(self, arg1="", arg2=0):
    self.name =arg1
    self.number =arg2
    self.total_unit = 0
    self.judge = ""

  def set_name(self, arg1):
    if arg1 != "":
      self.name = arg1

  def set_unit(self, arg1):
    self.total_unit = arg1
    return(self.total_unit)

  def set_number(self, arg1):
    self.number = arg1

  def disp(self):
    print("学生番号:{}, 氏名:{}".format(self.number, self.name))
    print("総単位:{}".format(self.total_unit))
    print("進級判定:{}".format(self.judge))

  def judgement_process(self):
    if self.total_unit >= 60:
      self.judge = "進級"
    else:
      self.judge = "残念"

shin = student("しん")
shin.set_number("0001")
shin.set_unit(50)
shin.judgement_process()
shin.disp()

yan = student("やん")
yan.set_number("0002")
yan.set_unit(70)
yan.judgement_process()
yan.disp()
```
