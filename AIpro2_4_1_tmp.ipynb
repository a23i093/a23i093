# 課題1 <br>
0～9までの中から整数値をinputで入力し、<br>
その整数値をラベルとした画像を3つ出力せよ。

```python
import matplotlib.pyplot as plt
from sklearn import datasets

digits_data = datasets.load_digits()
size=digits_data.data.shape[0]

n=int(input("0～9の間の整数値:"))
L_num=3  #出力する個数の設定
nn=0

#0からデータサイズ-1までチェックする
for i in range(size):
  if(digits_data.target[i] == n):
    nn +=1
    plt.figure(figsize=(5,2))
    print("ラベル:", digits_data.target[i])
    plt.imshow(digits_data.data[i].reshape(8,8))
    plt.show()
    #表示個数3を超えたらbreakする。
    if(nn >= L_num):
      break;
```

# 課題2<br>
0～9のデータそれぞれの個数を数えて出力しなさい。

```python
import matplotlib.pyplot as plt
from sklearn import datasets

#データの読み込み
digits_data = datasets.load_digits()
#データの個数の設定
size=digits_data.data.shape[0]

#0～9までの頻度を入れておく辞書データ
#n_num[0]なら0の出現頻度が入る
n_num={}

#初期化(0～9までの頻度を0に)
for i in range(10):
  n_num[i]=0

#データ全部をみて、数字の出現個数を数える
for i in range(size):
  #正解ラベルのデータを添え字にして、1加える
  n_num[digits_data.target[i]] += 1

#出力処理
for i in range(10):
  print(i,"は",n_num[i],"個あります。")
```
