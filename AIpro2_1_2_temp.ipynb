(1)合計sumと平方根sqrt<br>
ベクトルの大きさの計算

```python
L= [1,3,5,2,4,6,2,3,7]
N=np.array(L)

#各要素の自乗
N2=N*N
print(N2)

#自乗した後、各要素を合計
print(np.sum(N2))
#自乗した後、各要素を合計したものを平方根をとる。
print("Lのベクトルの大きさ:", np.sqrt(np.sum(N2)))
```
