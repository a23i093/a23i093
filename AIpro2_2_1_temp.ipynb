#4.numpyで行列の計算(その２）

(1)逆行列

```python
W3=[[1,2],[4,5]]

MW3=np.array(W3)

#逆行列の計算
MW4 = np.linalg.inv(MW3)
#MW4 = MW3.I

print("MW3\n",MW3)
print("MW4\n",MW4)

#Xは単位行列になるはず
X=np.dot(MW3, MW4)
print("X\n",X)
```

#課題1<br>
連立方程式の解の計算<br>
y=Mx → x=(Mの逆行列) y


```python
M=np.array([[1,2],[3,1]])
y=np.array([1,1])

#逆行列の計算
MI=np.linalg.inv(M)

x=np.dot(MI,y)

print("連立方程式の解:\n",x)
```

#課題2<br>
MA・x + bの計算結果の正規化せよ。<br>
MA・x + bの計算のベクトルの大きさを計算した後、<br>
各要素を大きさで割ればよい。<br>
プログラムはnp.linalg.normは使わないで、<br>
上セルを組み合わせる

```python
#前半は前回の課題2と同じ
x=[1,0,1]
x=np.array(x)

W1=[[1,2,3],[4,5,6],[-1,0,7]]
MA=np.array(W1)

b=3
I=np.full(3,1)

Y=np.dot(MA,x)+np.dot(I,b)
print("計算結果",Y)

#後半は、ベクトルの大きさを求める。
#各要素の2乗
N2=Y*Y
print(N2)
#要素の2乗の総和の平方根　大きさを計算
SIZE=np.sqrt(np.sum(N2))
#ベクトルの各要素を大きさで割る
Y1=[i for i in Y/SIZE]

print("Y1:",Y1)
```

#6.numpyで提供されている関数

(1)合計sumと平方根sqrt<br>
ベクトルの大きさの計算

```python
L= [1,3,5,2,4,6,2,3,7]
N=np.array(L)

#各要素の2乗
N2 =N*N
print(N2)

#2乗した後、各要素を合計
N2_sum=np.sum(N2)
print(N2_sum)
#2乗した後、各要素を合計した
#ものを平方根sqrtをとる。
N2_sum_sqrt=np.sqrt(N2_sum)
print("Lのベクトルの大きさ:", N2_sum_sqrt)
```

(1)ReLU関数<br>

```python
import numpy as np
import matplotlib.pyplot as plt

#0.1刻み
X= np.arange(-10, 10, 0.1)
Y1= np.maximum(0, X)

plt.plot(X,Y1)
plt.show()
```
